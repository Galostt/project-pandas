```python
def eliminar_columnas_y_filas(dataframe):
    eliminar_cols_y_fils = dataframe.copy()
    eliminar_cols_y_fils.dropna(axis=1, how='all', inplace= True)
    eliminar_cols_y_fils.dropna(axis=0, how='all', inplace= True)
    return eliminar_cols_y_fils
```

```python
def eliminar_columnas_rep(dataframe2):
    elimnar_cols_rep = dataframe2.copy()
    elimnar_cols_rep = elimnar_cols_rep[['Date', 'Type', 'Country', 'Area', 'Location',
                                        'Activity', 'Name', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Time',
                                        'Species ', 'Investigator or Source', 'pdf', 'href formula','original order']]
    return elimnar_cols_rep
```

```python
def eliminar_espacios_en_titulos(dataframe3):
    elimina_espacios = dataframe3.copy()
    elimina_espacios.columns = elimina_espacios.columns.str.strip()
    return elimina_espacios
```

```python
def Cambiar_orden_de_columnas(dataframe4):
    ordenar_cols = dataframe4.copy()
    ordenar_cols = ordenar_cols[['original order','Name','Sex', 'Age', 'Injury', 'Fatal (Y/N)','Activity',
                                 'Type','Date','Country', 'Area', 'Location', 'Species', 'Investigator or Source', 
                                 'pdf', 'href formula']]
    return ordenar_cols
```

```python
def Cambiar_nombre_de_columnas(dataframe5):
    cambiar_nombre = dataframe5.copy()
    cambiar_nombre = cambiar_nombre.rename(columns ={'original order':'Order','Name':'FullName','Investigator or Source':'Source','pdf':'Image',
                                                     'href formula':'href'})
    return cambiar_nombre
```

```python
def Arreglo_de_order(dataframe6):
    cambio_order = dataframe6.copy()
    cambio_order['Order'] = cambio_order['Order'].fillna(0).astype(np.int64, errors = 'ignore')
    
    return cambio_order
```

```python
def eliminar_vacios(dataframe7):
    eliminar_nans = dataframe7.copy()
    columnas = ['Order', 'FullName', 'Sex', 'Age', 'Injury', 'Fatal (Y/N)', 'Activity','Type', 
                'Date', 'Country', 'Area', 'Location', 'Species', 'Source','Image', 'href']
    for columna in columnas:
        eliminar_nans[columna] = eliminar_nans[columna].fillna('Unknown')
    eliminar_nans = eliminar_nans.fillna(0)

    return eliminar_nans
```

```python
def eliminar_renglones_vacios(dataframe8):
    eliminar_rows = dataframe8.copy()
    eliminar_rows = eliminar_rows[eliminar_rows['Date']!='Unknown']
    eliminar_rows = eliminar_rows.drop_duplicates()
    eliminar_rows = eliminar_rows.sort_values(by=['Order']).reset_index()
    return eliminar_rows
```

```python
def definition_code_type(dataframe9):
    code_type = dataframe9.copy()
    code_type['Encode_Type'] = code_type['Type'].apply(definition_type)
    return code_type
```

```python
def definition_type(list1):
    if list1 == 'Unknown':
        return 0
    if list1 == 'Unprovoked':
        return 1
    if list1 == 'Boat':
        return 2
    if list1 == 'Sea Disaster':
        return 3
    if list1 == 'Provoked':
        return 4
    if list1 == 'Invalid':
        return 5
    if list1 == 'Boating':
        return 6
    if list1 == 'Boatomg':
        return 7
    if list1 == 'Questionable':
        return 8
```

```python
data1 = eliminar_columnas_y_filas(attacks2)
data2 = eliminar_columnas_rep(data1)
data3 = eliminar_espacios_en_titulos(data2)
data4 = Cambiar_orden_de_columnas(data3)
data5 = Cambiar_nombre_de_columnas(data4)
data6 = Arreglo_de_order(data5)
data7 = eliminar_vacios(data6)
data8 = eliminar_renglones_vacios(data7)
data9 = definition_code_type(data8)
data9.to_csv('main.csv')
```
